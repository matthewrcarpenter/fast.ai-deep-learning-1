## Load daily data

```python
ticker = 'IAU'
daily = fat.get_price_data(ticker)
```

```python
daily.tail()
```

```python
daily['Adj Close'].dropna().plot()
```

## Resample to Weekly

```python
weekly = daily.resample('W', loffset=pd.offsets.timedelta(days=-7)).mean()['Adj Close']
```

```python
weekly.tail()
```

```python
path = f'data/google/{ticker}-weekly-mean_close.csv'
weekly.to_csv(path)
```

```python
weekly.plot()
```

```python
negative_weekly = -weekly
```

```python
negative_weekly.plot()
```

```python
path = f'data/google/{ticker}-negative-weekly-mean_close.csv'
negative_weekly.to_csv(path)
```
